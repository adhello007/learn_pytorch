Random Tensors 

Random tensors are important because many neural networks start with tensors full of random numbers and then adjust those random numbers 
to better represent the data. 

```python
rand = torch.rand(3,4)
```

```python
rand
```

```python
random_image = torch.rand(size=(3,512,512)) 
```

```python
random_image.shape
```

```python
import numpy as np 
im = random_image.numpy()
```

```python
im_transpose = np.transpose(im,(1,2,0)) #0 means channel, 1 means height, 2 means width 
```

```python
im_final_np  = (im_transpose*255).astype(np.uint8)
```

```python
from PIL import Image 
image = Image.fromarray(im_final_np)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```
